 4)CONVERTING THE TARGET VARIABLE CHANCE OF ADMIT TO CATEGORICAL VALUES 0 AND 1 



```python
file.loc[file['Chance of Admit ']<=0.80, 'Chance of Admit']="0"
file.loc[file['Chance of Admit ']>=0.80, 'Chance of Admit']="1"
  
```

```python
x=file.drop(columns=(['Serial No.','GRE Score','TOEFL Score','University Rating','SOP','CGPA','Research']))
x
```

```python
y=file["Chance of Admit"]
```

                         SPLITTING THE SETS INTO X AND Y

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=10)
```

```python
print(xtrain,ytrain)
```

                  USING LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
model.predict_proba(xtest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(ytest,ypred)
```

```python
confusion_matrix(ytest,ypred)
```

```python
print(classification_report(ytest,ypred))
```

              5)USING DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
model.predict_proba(xtest)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(ytest,ypred)
```

```python
confusion_matrix(ytest,ypred)
```

```python
print(classification_report(ytest,ypred))
```

Here in Decision Tree,we got  higher accuacy without using hyper parameter tunings.

            6)USING RANDOM FOREST CLASSIFIERS

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=100)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rand_f = RandomForestClassifier(n_estimators = 200, criterion = 'entropy', 
                                min_samples_split = 10, min_samples_leaf = 9, max_features = "auto",
                                random_state = 500, max_depth = 12) 
```

```python
rf_fit=rand_f.fit(x_train,y_train)
rf_fit
```

```python
y_pred=rf_fit.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
